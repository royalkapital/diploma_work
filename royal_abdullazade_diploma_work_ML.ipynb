```python
import json
import pandas as pd
from sklearn.ensemble import IsolationForest


FILE_PATH = '/content/drive/MyDrive/ecar_dataset.csv'
CHUNKSIZE = 50000
SAMPLE_PER_CHUNK = 500


def preprocess_chunk(chunk: pd.DataFrame, fit_cols: list = None) -> pd.DataFrame:


    if not chunk.empty:
        header_mask = chunk.apply(
            lambda col: col.astype(str).str.strip() == col.name, axis=0
        ).all(axis=1)
        chunk = chunk[~header_mask].copy()

    def _safe_json_parse(raw):
        if not isinstance(raw, str):
            return {}
        try:
            parsed = json.loads(raw)
            return parsed if isinstance(parsed, dict) else {}
        except:
            return {}

    chunk['properties'] = chunk['properties'].apply(_safe_json_parse)


    prop_df = pd.json_normalize(chunk['properties'])
    prop_df.index = chunk.index

    chunk = pd.concat([chunk.drop(columns=['properties']), prop_df], axis=1)


    drop_cols = ['pid', 'ppid', 'tid', 'id', 'objectID', 'actorID']
    chunk = chunk.drop(columns=[c for c in drop_cols if c in chunk.columns])


    cat_cols = ['action', 'hostname', 'object', 'principal']
    for col in cat_cols:
        if col in chunk.columns:
            chunk[col] = chunk[col].astype('category').cat.codes

    # Timestamp features
    if 'timestamp' in chunk.columns:
        chunk['timestamp'] = pd.to_datetime(chunk['timestamp'], errors='coerce')
        chunk['hour'] = chunk['timestamp'].dt.hour
        chunk['minute'] = chunk['timestamp'].dt.minute
        chunk['weekday'] = chunk['timestamp'].dt.weekday


        chunk = chunk.drop(columns=['timestamp'])


    if 'command_line' in chunk.columns:
        chunk['cmd_length'] = chunk['command_line'].astype(str).apply(len)
    else:
        chunk['cmd_length'] = 0


    chunk = chunk.fillna(0)


    if fit_cols is not None:
        for col in fit_cols:
            if col not in chunk.columns:
                chunk[col] = 0
        chunk = chunk[fit_cols]

    return chunk


def fit_model(file_path, sample_per_chunk, chunksize):
    sample_frames = []
    fit_cols = None

    for chunk in pd.read_csv(file_path, chunksize=chunksize, low_memory=False):
        chunk = preprocess_chunk(chunk)

        numeric_cols = chunk.select_dtypes(include='number').columns

        if fit_cols is None:
            fit_cols = numeric_cols

        chunk_sample = chunk[fit_cols].sample(
            n=min(sample_per_chunk, len(chunk)),
            random_state=42
        )

        sample_frames.append(chunk_sample)

    fit_data = pd.concat(sample_frames, ignore_index=True)

    model = IsolationForest(
        n_estimators=100,
        contamination=0.1,
        random_state=42,
        n_jobs=-1
    )

    model.fit(fit_data)

    print(f"Model fit completed. Fit data shape: {fit_data.shape}")
    return model, fit_cols


def predict_anomalies(file_path, model, fit_cols, chunksize):
    results = []

    for chunk in pd.read_csv(file_path, chunksize=chunksize, low_memory=False):
        chunk = preprocess_chunk(chunk, fit_cols=fit_cols)

        chunk_features = chunk[fit_cols]

        chunk['anomaly'] = model.predict(chunk_features)
        chunk['anomaly'] = chunk['anomaly'].map({1: 0, -1: 1})

        chunk['anomaly_score'] = model.decision_function(chunk_features)

        results.append(chunk[['anomaly', 'anomaly_score']])

    df_result = pd.concat(results, ignore_index=True)
    return df_result


if __name__ == "__main__":
    print("Fitting Isolation Forest model...")
    model, fit_cols = fit_model(FILE_PATH, SAMPLE_PER_CHUNK, CHUNKSIZE)

    print("Predicting anomalies on full dataset...")
    df_result = predict_anomalies(FILE_PATH, model, fit_cols, CHUNKSIZE)

    print(f"Final result shape: {df_result.shape}")
    print(f"Number of anomalies detected: {df_result['anomaly'].sum()}")
    print(df_result.head(10))
```

```python
def label_true_false(df):

    suspicious_keywords = [
        "powershell",
        "cmd.exe",
        "wget",
        "curl",
        "enc",
        "base64"
    ]

    def check_tp(cmd):

        if not isinstance(cmd, str):
            return 0

        cmd = cmd.lower()

        for keyword in suspicious_keywords:
            if keyword in cmd:
                return 1

        return 0

    df['true_positive'] = df[
        'command_line'
    ].apply(check_tp)

    df['false_positive'] = 1 - df[
        'true_positive'
    ]

    return df
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import f1_score, precision_score, recall_score
import ast
import time
import numpy as np
import warnings
warnings.filterwarnings('ignore')


FILE_PATH = '/content/drive/MyDrive/ecar_dataset.csv'
CHUNKSIZE = 30000
SAMPLE_PER_CHUNK = 20000
OUTPUT_FILE = '/content/drive/MyDrive/tp_fp_results.csv'

category_mappings = {}
CAT_COLS = ['action', 'hostname', 'object', 'principal']


SUSPICIOUS_KEYWORDS = [
    "powershell", "cmd.exe", "wget", "curl", "enc", "base64"
]



def safe_parse(x):
    try:
        if pd.isna(x):
            return {}
        if isinstance(x, str):
            return ast.literal_eval(x)
        return {}
    except:
        return {}


def extract_properties(chunk):
    chunk = chunk.copy()

    if 'properties' in chunk.columns:
        chunk['properties'] = chunk['properties'].apply(safe_parse)

        chunk['command_line'] = chunk['properties'].apply(
            lambda x: x.get('command_line', '') if isinstance(x, dict) else ''
        )
        chunk['image_path'] = chunk['properties'].apply(
            lambda x: x.get('image_path', '') if isinstance(x, dict) else ''
        )

        chunk = chunk.drop(columns=['properties'])

    return chunk



def detect_keywords(text):

    text_lower = str(text).lower()
    return 1 if any(k in text_lower for k in SUSPICIOUS_KEYWORDS) else 0



def create_features(chunk):
    chunk = chunk.copy()

    if 'command_line' in chunk.columns:
        cmd = chunk['command_line'].astype(str)
        cmd_lower = cmd.str.lower()


        chunk['keyword_match'] = cmd.apply(detect_keywords)


        chunk['has_powershell'] = cmd_lower.str.contains('powershell', na=False).astype(int)
        chunk['has_cmd_exe'] = cmd_lower.str.contains('cmd.exe', na=False).astype(int)
        chunk['has_wget'] = cmd_lower.str.contains('wget', na=False).astype(int)
        chunk['has_curl'] = cmd_lower.str.contains('curl', na=False).astype(int)
        chunk['has_enc'] = cmd_lower.str.contains('enc', na=False).astype(int)
        chunk['has_base64'] = cmd_lower.str.contains('base64', na=False).astype(int)

        chunk['keyword_count'] = (
            chunk['has_powershell'] + chunk['has_cmd_exe'] +
            chunk['has_wget'] + chunk['has_curl'] +
            chunk['has_enc'] + chunk['has_base64']
        )


        chunk['cmd_length'] = cmd.apply(len)
        chunk['cmd_word_count'] = cmd.apply(lambda x: len(x.split()))
        chunk['cmd_special_chars'] = cmd.apply(lambda x: sum(1 for c in x if c in '|&><;$`\\(){}[]'))
        chunk['cmd_dash_count'] = cmd.apply(lambda x: x.count('-'))
        chunk['cmd_slash_count'] = cmd.apply(lambda x: x.count('\\'))


        chunk['has_hidden'] = cmd_lower.str.contains('hidden', na=False).astype(int)
        chunk['has_bypass'] = cmd_lower.str.contains('bypass', na=False).astype(int)
        chunk['has_download'] = cmd_lower.str.contains('download', na=False).astype(int)
        chunk['has_invoke'] = cmd_lower.str.contains('invoke', na=False).astype(int)
        chunk['has_execution'] = cmd_lower.str.contains('execution', na=False).astype(int)


        def calc_entropy(s):
            if len(s) == 0:
                return 0
            prob = [float(s.count(c)) / len(s) for c in set(s)]
            return -sum([p * np.log2(p) for p in prob if p > 0])

        chunk['cmd_entropy'] = cmd.apply(calc_entropy)


        chunk['has_http'] = cmd_lower.str.contains('http', na=False).astype(int)
        chunk['has_exe'] = cmd_lower.str.contains('.exe', na=False).astype(int)


        chunk['suspicion_score'] = (
            chunk['keyword_count'] * 2 +
            chunk['has_hidden'] + chunk['has_bypass'] +
            chunk['has_download'] + chunk['has_invoke'] +
            chunk['has_http']
        )


    if 'image_path' in chunk.columns:
        img = chunk['image_path'].astype(str).str.lower()
        chunk['is_system32'] = img.str.contains('system32', na=False).astype(int)
        chunk['is_temp'] = img.str.contains('temp', na=False).astype(int)
        chunk['is_windows'] = img.str.contains('windows', na=False).astype(int)

    return chunk



def encode_categorical(chunk, col, is_training=False):
    global category_mappings

    chunk[col] = chunk[col].astype(str).fillna('unknown')

    if is_training:
        if col not in category_mappings:
            category_mappings[col] = {}
        for val in chunk[col].unique():
            if val not in category_mappings[col]:
                category_mappings[col][val] = len(category_mappings[col])

    if col in category_mappings:
        chunk[col] = chunk[col].map(category_mappings[col]).fillna(-1).astype(int)
    else:
        chunk[col] = 0

    return chunk



def preprocess_for_model(chunk, fit_cols=None, is_training=False):
    chunk = chunk.copy()

    drop_cols = ['pid', 'ppid', 'tid', 'id', 'objectID', 'actorID',
                 'command_line', 'image_path']
    chunk = chunk.drop(
        columns=[c for c in drop_cols if c in chunk.columns],
        errors='ignore'
    )

    for col in CAT_COLS:
        if col in chunk.columns:
            chunk = encode_categorical(chunk, col, is_training)

    if 'timestamp' in chunk.columns:
        chunk['timestamp'] = pd.to_datetime(chunk['timestamp'], errors='coerce')
        chunk['hour'] = chunk['timestamp'].dt.hour.fillna(0).astype(int)
        chunk['minute'] = chunk['timestamp'].dt.minute.fillna(0).astype(int)
        chunk['weekday'] = chunk['timestamp'].dt.weekday.fillna(0).astype(int)
        chunk = chunk.drop(columns=['timestamp'])

    chunk = chunk.fillna(0)
    numeric_cols = chunk.select_dtypes(include=['number']).columns.tolist()
    chunk = chunk[numeric_cols].astype(float)

    if fit_cols is not None:
        for col in fit_cols:
            if col not in chunk.columns:
                chunk[col] = 0.0
        chunk = chunk[fit_cols]

    return chunk



def fit_model():
    global category_mappings
    category_mappings = {}

    sample_frames = []
    label_frames = []
    fit_cols = None
    chunk_count = 0

    print("=" * 60)
    print("TRAINING PHASE ")
    print("=" * 60)

    for chunk in pd.read_csv(
        FILE_PATH,
        chunksize=CHUNKSIZE,
        low_memory=False
    ):
        chunk_count += 1

        chunk = extract_properties(chunk)
        chunk = create_features(chunk)

        labels = chunk['keyword_match'].copy()

        processed = preprocess_for_model(chunk, is_training=True)

        if fit_cols is None:
            fit_cols = processed.columns.tolist()
            print(f"Features ({len(fit_cols)}): {fit_cols}")

        for col in fit_cols:
            if col not in processed.columns:
                processed[col] = 0.0
        processed = processed[fit_cols]

        processed['label'] = labels.values[:len(processed)]


        normal = processed[processed['label'] == 0]
        anomaly = processed[processed['label'] == 1]

        n = min(SAMPLE_PER_CHUNK // 2, len(normal), len(anomaly))

        if n > 100:
            normal_sample = normal[fit_cols].sample(n=n, random_state=42)
            anomaly_sample = anomaly[fit_cols].sample(n=min(n, len(anomaly)), random_state=42)

            sample_frames.append(pd.concat([normal_sample, anomaly_sample], ignore_index=True))
            label_frames.append(pd.Series([0]*len(normal_sample) + [1]*len(anomaly_sample)))

        print(f"Chunk {chunk_count}: Balanced samples = {n*2 if n > 100 else 0}")

        if chunk_count == 120:
            break

    fit_data = pd.concat(sample_frames, ignore_index=True).astype(float)
    fit_labels = pd.concat(label_frames, ignore_index=True).astype(int)

    print(f"\nTotal training samples: {len(fit_data)}")
    print(f"Label distribution: {fit_labels.value_counts().to_dict()}")

    scaler = RobustScaler()
    fit_data_scaled = scaler.fit_transform(fit_data)


    print("\nTraining Gradient Boosting Classifier...")
    model = GradientBoostingClassifier(
        n_estimators=200,
        max_depth=8,
        min_samples_split=50,
        min_samples_leaf=25,
        learning_rate=0.1,
        subsample=0.8,
        random_state=42
    )
    model.fit(fit_data_scaled, fit_labels)


    train_proba = model.predict_proba(fit_data_scaled)[:, 1]


    print("\nFinding optimal threshold...")
    best_threshold = 0.5
    best_f1 = 0

    for thresh in np.arange(0.3, 0.8, 0.02):
        preds = (train_proba >= thresh).astype(int)
        f1 = f1_score(fit_labels, preds, zero_division=0)
        prec = precision_score(fit_labels, preds, zero_division=0)
        rec = recall_score(fit_labels, preds, zero_division=0)


        if f1 > best_f1:
            best_f1 = f1
            best_threshold = thresh
            print(f"  Threshold {thresh:.2f}: P={prec:.4f}, R={rec:.4f}, F1={f1:.4f}")

    print(f"\nOptimal threshold: {best_threshold:.2f} (F1: {best_f1:.4f})")


    final_preds = (train_proba >= best_threshold).astype(int)
    print(f"\nTraining Metrics at threshold {best_threshold:.2f}:")
    print(f"  Precision: {precision_score(fit_labels, final_preds):.4f}")
    print(f"  Recall:    {recall_score(fit_labels, final_preds):.4f}")
    print(f"  F1-Score:  {f1_score(fit_labels, final_preds):.4f}")

    print("\nModel ready!")
    return model, fit_cols, scaler, best_threshold



def predict_and_save(model, fit_cols, scaler, threshold):
    first = True
    total_chunks = 0
    total_anomalies = 0
    start_time = time.time()

    print("\n" + "=" * 60)
    print("PREDICTION PHASE (Smart Hybrid)")
    print(f"ML Threshold: {threshold:.2f}")
    print("=" * 60)

    for chunk in pd.read_csv(
        FILE_PATH,
        chunksize=CHUNKSIZE,
        low_memory=False
    ):
        total_chunks += 1

        chunk_extracted = extract_properties(chunk.copy())
        chunk_extracted = create_features(chunk_extracted)

        output_cols = ['action', 'hostname', 'object', 'principal',
                       'timestamp', 'command_line', 'image_path']
        output_chunk = chunk_extracted[[c for c in output_cols if c in chunk_extracted.columns]].copy()


        keyword_match = chunk_extracted['keyword_match'].values


        processed = preprocess_for_model(chunk_extracted.copy(), fit_cols, is_training=False)
        features = scaler.transform(processed.astype(float).fillna(0))
        ml_proba = model.predict_proba(features)[:, 1]



        final_preds = np.where(
            (keyword_match == 1) & (ml_proba >= threshold),
            1,
            0
        )

        output_chunk = output_chunk.reset_index(drop=True)
        output_chunk['anomaly'] = final_preds
        output_chunk['anomaly_score'] = ml_proba

        chunk_anomalies = output_chunk['anomaly'].sum()
        total_anomalies += chunk_anomalies

        output_chunk.to_csv(
            OUTPUT_FILE,
            mode='a',
            index=False,
            header=first
        )
        first = False

        elapsed = int(time.time() - start_time)
        print(f"Chunk {total_chunks}: {chunk_anomalies} anomalies | Time: {elapsed}s")

    print("=" * 60)
    print(f"COMPLETE! Total anomalies: {total_anomalies}")
    print("=" * 60)



if __name__ == "__main__":

    print("\n" + "=" * 70)
    print("  Done")
    print("  Done ")
    print("=" * 70 + "\n")

    model, fit_cols, scaler, threshold = fit_model()
    predict_and_save(model, fit_cols, scaler, threshold)

    print("\n DONE!")
```

```python
import pandas as pd

FILE_PATH = '/content/drive/MyDrive/tp_fp_results.csv'
CHUNKSIZE = 50000

suspicious_keywords = [
    "powershell", "cmd.exe", "wget", "curl", "enc", "base64"
]

TP = 0
FP = 0
TN = 0
FN = 0

for chunk in pd.read_csv(FILE_PATH, chunksize=CHUNKSIZE, low_memory=False):
    if 'anomaly' not in chunk.columns:
        continue

    if 'command_line' in chunk.columns:
        chunk['true_label'] = chunk['command_line'].astype(str).apply(
            lambda x: 1 if any(k in x.lower() for k in suspicious_keywords) else 0
        )
    else:
        chunk['true_label'] = 0

    chunk['pred_label'] = chunk['anomaly']

    for t, p in zip(chunk['true_label'], chunk['pred_label']):
        if t == 1 and p == 1:
            TP += 1
        elif t == 0 and p == 1:
            FP += 1
        elif t == 0 and p == 0:
            TN += 1
        elif t == 1 and p == 0:
            FN += 1

print("=" * 50)
print("CONFUSION MATRIX")
print("=" * 50)
print(f"[[TN={TN}, FP={FP}]")
print(f" [FN={FN}, TP={TP}]]")

total = TP + FP + TN + FN
accuracy = (TP + TN) / total
precision = TP / (TP + FP) if (TP + FP) > 0 else 0
recall = TP / (TP + FN) if (TP + FN) > 0 else 0
f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

print()
print("=" * 50)
print("METRICS")
print("=" * 50)
print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1-Score:  {f1:.4f}")
```
